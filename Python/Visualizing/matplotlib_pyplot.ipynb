### 折线图plt.plot()

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
# 创建一幅线图，x轴是年份，y轴是gdp
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
# 添加一个标题
plt.title("名义GDP")
# 给y轴加标记
plt.ylabel("十亿美元")
plt.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]

xs = [i for i, _ in enumerate(variance)]  #enumerate(list)返回list的index和对应的内容

# 可以多次调用plt.plot，以便在同一个图上显示多个序列
plt.plot(xs, variance, 'g-', label='variance') # 绿色实线
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # 红色点虚线
plt.plot(xs, total_error, 'b:', label='total error') # 蓝色点线

# 因为已经对每个序列都指派了标记,所以可以自由地布置图例
plt.legend(loc=9) # loc=9指的是“顶部中央”
plt.xlabel("模型复杂度")
plt.title("偏差-方差权衡图")
plt.show()
```

### 条形图plt.bar()

```python
#数据
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
# 使用左侧x坐标[xs]和高度[num_oscars]画条形图
xs = [i for i, _ in enumerate(movies)]
plt.bar(xs, num_oscars)
plt.ylabel("所获奥斯卡金像奖数量")
plt.title("我最喜爱的电影")
# 给x轴添加label
plt.xticks([i for i, _ in enumerate(movies)], movies)
plt.show()
```

### 散点图plt.scatter()

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)
# 每个点加标记
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, #注释内容
                 xy=(friend_count, minute_count), # 把标记放在对应的点上
                 xytext=(5, -5), # 但要有轻微偏离
                 textcoords='offset points')
plt.title("日分钟数与朋友数")
plt.xlabel("朋友数")
plt.ylabel("花在网站上的日分钟数")
plt.show()
```
