```python
def get_info(xpath):
    
    global info
    
    element = driver.find_element_by_xpath(xpath)
    ActionChains(driver).move_to_element(element).click().perform()
    driver.switch_to.window(driver.window_handles[-1])
    driver.switch_to.window(driver.window_handles[-1])
    time.sleep(7)
    
    name = driver.find_element_by_xpath('/html/body/div/div[3]/div/div/h3').text 
    company = driver.find_element_by_xpath('/html/body/div/div[4]/div[2]/div[1]/div/a[1]').text
    salary = driver.find_element_by_class_name('summary-plane__salary').text
    edu = driver.find_element_by_xpath('/html/body/div/div[3]/div/div/div[2]/div[1]/ul/li[3]').text
    detail = driver.find_element_by_class_name('describtion__detail-content').text
    
    treat0 = driver.find_elements_by_class_name('highlights__content') #某些公司没有员工福利信息
    if len(treat0)>0:
        treat = treat0[0].text
    else:
        treat = ''
    
    total = {'name':name,'company':company,'salary':salary,'edu':edu,'treat':treat,'detail':detail}
    info = info.append(pd.DataFrame(total,index=range(i,i+1)))
    driver.close()
    driver.switch_to.window(mainhandle)
```

```python
info = pd.DataFrame()
```

```python
for i in range(1,91):
    xpath = '/html/body/div/div/div[4]/div[3]/div[2]/div/div[%s]/div/a/div[1]/div[1]/span'%(i)
    js = "window.scrollTo(0,%s)"%(i*142)  #滑动窗口到指定位置
    driver.execute_script(js)
    time.sleep(2)
    get_info(xpath)
```
